```python
# Importar librerías

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Leer la DB

df = pd.read_csv(r'/Users/danna/Desktop/Data Science - Python/Movies portfolio project/movies.csv')
```

```python
# Visualizar la DB

df
```

```python
# Ver si hay datos nulos

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Ver tipos de datos

print(df.dtypes)
```

```python
# Descartar lineas duplicadas

df.drop_duplicates()
```

```python
# Ordenar DB por 'gross' de mayor a menor

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Representar relación busget-gross gráfiamente

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for firm')
plt.show()
```

```python
# Representar relación budget-gross gráfiamente y mostrando correlación (regression plot)

sns.regplot(x="gross", y="budget", data=df, line_kws={"color":"b"})
```

```python
# Ver matriz de correlación pearson
df.corr(method='pearson')
```

```python
# Ver matriz de correlación kendall
df.corr(method='kendall')
```

```python
# Ver matriz de correlación spearman
df.corr(method='spearman')
```

```python
# Visualizar gráficamente matriz de correlación pearson (heat map)
matriz_corr = df.corr(method='pearson')
sns.heatmap(matriz_corr, annot=True)
plt.title('Matriz de correlación para datos numéricos')
plt.show()
```

```python
# Asignar valores numéricos a los datos que eran type object para incluirlos en el heat map
df_ennum = df
for col_name in df_ennum.columns:
    if df_ennum[col_name].dtype == 'object':
        df_ennum[col_name] = df_ennum[col_name].astype('category')
        df_ennum[col_name] = df_ennum[col_name].cat.codes
df_ennum
```

```python
# Visualizar gráficamente matriz de correlación pearson (heat map)
matriz_corr = df_ennum.corr(method='pearson')
sns.heatmap(matriz_corr, annot=True)
plt.title('Matriz de correlación para datos numéricos')
plt.show()
```

```python
# Visualizo la matriz de correlación
df_ennum.corr()
```

```python
matriz_corr = df_ennum.corr()
corr_pairs = matriz_corr.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Mostrar combinaciones con coreelación alta (> 0.5)
# No cuentas los valores iguales a 1, ya que es la correlación de un dato con si mismo
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr 
```
